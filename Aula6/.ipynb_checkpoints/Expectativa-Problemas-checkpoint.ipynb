```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('expectativa-completo.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
```

```python
df = df.loc[:, ~(df.isna().sum() > 70)]
```

```python
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = df.loc[:, df.columns != "2020"]
y = df['2020']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=0)
```

```python
numerical_transform = Pipeline(steps = [('simple_imputer', SimpleImputer(strategy='median')),
                                                                        ('minmax_scaler', MinMaxScaler())])
```

```python
preprocessor = ColumnTransformer(transformers = [('num', numerical_transform, X.select_dtypes(exclude="object").columns)])
```

```python
lrmodel = LinearRegression()
```

```python
pipe = Pipeline(steps=[('preprocessor', preprocessor), ('lr_model', lrmodel)])
```

```python
pipe.fit(X_train, y_train)
```
