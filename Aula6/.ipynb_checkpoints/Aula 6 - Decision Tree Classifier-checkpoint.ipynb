Finally, let's attempt to predict the BAD clients using a DecisionTree model.

# Model Implementation - Decision Tree

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
X = df.loc[:, df.columns != "BAD"]
y = df.BAD

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
```

```python
X_train.professionCode = X_train.professionCode.astype(object)
X_test.professionCode = X_test.professionCode.astype(object)
```

```python
scaler = StandardScaler()
```

```python
encoder = OneHotEncoder(handle_unknown='ignore')
```

```python
preprocessor = ColumnTransformer(transformers = [('cat', encoder, X.select_dtypes(include="object").columns),
                                                 ('num', scaler, X.select_dtypes(exclude="object").columns)])
```

```python
dtcmodel = DecisionTreeClassifier(random_state=0, max_depth=6)
```

```python
pipe = Pipeline(steps=[('preprocessor', preprocessor), ('dtc_model', dtcmodel)])
```

```python
pipe.fit(X_train, y_train)
```

```python
preds = pipe.predict(X_test)
```

# Evaluate the model

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score, f1_score, classification_report, accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
print('Accuracy: ', accuracy_score(y_test, preds))
print('Precision: ', precision_score(y_test, preds))
print('Recall: ', recall_score(y_test, preds))
print('F1 Score: ', f1_score(y_test, preds))
print('\nClassification report:\n', classification_report(y_test, preds))
```

```python
cm = confusion_matrix(y_test, preds, labels=pipe['dtc_model'].classes_)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=pipe['dtc_model'].classes_)
```

```python
disp.plot()
plt.show()
```
